# Node.js 와 Express 로 벡엔드 입문하기 

2.1 Node.js 소개 
- 브라우저 밖에서 js 를 사용하는 v8 엔진을 사용. 
- 논블로킹/ 비동기 api 서버 환경구현 
- 논블로킹 감옥 : 동시 실행되는 스레드와 공유자원을 프로그래머가 직접 만들고 관리. 
- 비동기 환경 JSConf EU2009 에서 Node.js 런타임 환경을 처음 소개함. 
- 2010, npm 패키지 매니저 공개 , 2011 MS,조이언트사 도 윈도우 버전 node.js 출시 
- 가장 많이 사용되는 웹 프레임워크 기술 (익스프레스 3위)

2.2 Node.js 서버에서 어떻게 js 를 실행하는가?
- node.js 는 V8 js 엔진, libuv 및 C/C++ 에 의존성을 가진 js 런타임.

런타임이란? 
- 자바스크립트로 된 프로그램을 실행할수 있는 프로그램. 
- java 는 JRE (Java Runtime Envronment ), C# 은 CLR(Common Language Runtime)
- C언어는 런타임 없이 코드 실행하는 네이티브 언어임.
- 컴파일 과정을 마친 컴퓨터 프로그램이 실행되고 있는 환경 도는 동작되는 동안의 시간. 
- 우리가 이야기하는 런타임은 런타임 환경을 의미한다. 
- javascript 가 web browser 에서 실행되면 RE 는 Web browser 가 되고 Node.js 에서 동작하면 RE 는 Node.js 가 된다. 

런과 컴파일의 차이 
- 컴파일은 기계어 코드(자바의 경우 바이트코드)로 변환되어 하드웨어(자바의 경우 JVM)가 실행하게 하는 프로그램 
JVM(Java Virtual Machine)
컴파일 후에 실제로 동작하는것이 런 
키워드 : Node.js, V8엔진, libuv

2.2.2 JS 실행을 위한 V8 엔진 
- 엔진은 사용자가 작성한 코드를 실행하는 프로그램 
- 파서, 컴파일러, 인터프리터, 가비지 컬렉터, 콜 스택, 힙 으로 구성. 
- 인터프리터 역할을 하는 이그니션, 컴파일러 역할을 하는 터보펜 

콜스택이란?
- 현재 실행 중인 서브 루틴에 관한 정보 저장하는 스택 
힙 
- 객체나 동적 데이터가 저장되는 메모리 공간, 자료구조의 힙이 아님. 

JIT(just in time) compiler 
JS code -> 파서 -> AST(추상구문트리)-> 이그니션(최적화) -> 터보펜 -> Binary(binary code)


2.2.3 이벤트 루프와 운영체제 단 비동기 api 및 스레드 풀 지원 libuv
- Http, 파일, 소켓 통신 IO기능, JS 없는 기능 제공?
- libuv 라는 C++ 라이브러리 사용해 해결 
- 비동기 입출력, 이벤트 기반 라이브러리 
- 다양한 플랫폼에서 사용할 수잇는 이벤트 루프 제공 
    - epoll(linux), IOCP(window), kqueue(mac) emd. 
- 네트워크 파일 IO,DNS, 스레드 풀 기능 추가 제공 

2.2.4 Node.js 아키텍쳐 
- 런타임에 필요한 이벤트 루프 및 운영체제 시스템 api 사용시 libuv 라이브러리 사용함. 

이벤트 루프란?

자바스크립트 특징 
- 단일스레드 기반 언어로 한번에 하나의 작업만 처리 
- 비동기 동작이어서 단일 스레드임에도 동시에 많은 작업 수행
- 비동기 동작 핵심 요소는 js 가 아닌 브라우저가 가짐. 

브라우저 기본 구조 
- heap : 메모리 할당. 
- Call stack : 실행 코드 환경을 저장하는 자료구조. 함수호출시 저장됨. 가장 위에 쌓인 함수를 가장 먼저 처리. 
- Web APIs :브라우저 제공 api( DOM, Ajax,TimeOut 등.) callstack 이 실행한 비동기 함수가 Web api 호출, web api 는 콜백 함수를 task queue 에 넣음. 
- Callback Queue : 함수 저장 자료구조, 가장 먼저 들어온 함수 먼저 처리 
- Event Loop : Call stack 이 다 비워지면 callback queue 에 존재하는 함수를 하나씩 call stack 으로 옮김.
