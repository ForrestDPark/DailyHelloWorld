# REST API 
- 정의 : Http URI 로 잘 표현된 리소스에 대한 행위 HTTP Method, resoure 는 json, xml, yaml 등 표현 언어로 정의됨

- resource :  
- HTTPmethod : ( 행위, verb)
    - post, get, put, patch, delete  (CRUD 연산, creat, read, update, delete )
    - 리소스 정보 요청

- 표현 (Representaiton):
    POST http://www.plusblog.co.kr/user/newuser Content-Type: appllication/json
    {
        "username": "newusers",
        "age" : "20"
    }
    - HTTP 메소드의 정보 표현 부분을 메소드 바디 혹은 payload 라고 함. 

- REST API 특징, 장점. 
    1. Uniform : 리소스에 상관없이 동일한 api 메소드 사용
    2. Stateless : 서버측에서 수행의 문맥을 저장하지 않음. restapi 는 독립적처리함. 세션정보 서버유지 안함. 
    3. Cacheable : Http protocall 사용 을 위해 기존 웹 인프라 사용. 캐쉬 사용. 
    4. Self-Descriptiveness : 정보 내용 자체로 직관적 이해 가능. 
    5. Clinet-Server Architecture : rest api 서버 클라이언트에 api 제공 만 함. 
    6. Layerd System : multi layer 로 구성 될 수 있음. 
    7. Code on demand(Optional) :클라이언트 사이드 수행. 

- RestFull 하다. :
    1. 슬래시는 계층관계를 나타내는데 사용. / 로 끝나는것.은 안됨. 
    2. 가독성을위하 스네이크 표기는 자제. uri 는 소문자로 하는게 좋다. sql 은 대문자 해주면 좋음. 
    3. Users, users 는 다른 collection 임. 
    4. 파일 확장자는 uri에 포함시키지 않음. 
        : http://www.xxx.x.x/images/main.jpg -> http://www.xxx.x.x/images/main accept: image/jpg
    5. uri 에는 명사 위주로 쓰자. 
        GET http://x.x.x.x/users/delete-olduser ..
