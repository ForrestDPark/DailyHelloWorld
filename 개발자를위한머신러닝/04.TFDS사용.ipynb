```python
(training_images, training_labels),(test_images, test_labels)=\
    tfds.as_numpy(tfds.load('fashion_mnist',
                            split  =['train','test'], ## split 을 하지않으면 dic 반환,
                            ## split 변수를 지정하면 분할 세트의 리스트가 반환됨. 
                            batch_size = -1, ## tf.Tensor 타입의 전체 데이터 반화
                            as_supervised =True 
                            ))
def model_training_for_tfds(training_images, training_labels,test_images, test_labels):
    ## 2024.10.17
    ## fashion mnist 인 경우 tf.unit8 타입의 텐서가 반환되기떄문에  cast 를 통해 실수로 변환후 255로 나누어야함. 
    training_images = tf.cast(training_images,tf.float32)/255.0
    test_images = tf.cast(test_images,tf.float32)/255.0
    
    model = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(input_shape =(28,28,1)),
        tf.keras.layers.Dense(128, activation=tf.nn.relu),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(10,activation=tf.nn.softmax)
        
    ])
    model.compile(optimizer = 'adam',
                  loss = 'sparse_categorical_crossentropy',
                  metrics = ['accuracy'])
    model.fit(training_images,training_labels, epochs=5)
model_training_for_tfds(training_images, training_labels,test_images, test_labels)
```

```python
# 특성 디스크립션을 만들자. 
feature_description = {
    'image': tf.io.FixedLenFeature([],dtype=tf.string),
    'label': tf.io.FixedLenFeature([],dtype=tf.int64)
    
}

def _parse_function(example_proto):
    ## 위에서 만든 딕셔너리로 입력을 파싱
    return tf.io.parse_single_example(example_proto,feature_description)
parsed_dataset = raw_dataset.map(_parse_function)
for parsed_record in parsed_dataset.take(1):
    pprint((parsed_record))
```

## 4.6 텐서 플로에서 데이터 관리를 위한 ETL 프로세스 


```python
def ETL_process():
    #Extract(추출), Transform(변환), Load(적재)
    #https://f-lab.kr/insight/understanding-etl-process
    import tensorflow as tf
    import tensorflow_datasets as tfds
    ## model define
    conv_n =16
    conv_s = (3,3) 
    h_act = 'relu'
    mp_s = (2,2)
    model = tf.keras.models.Sequential([
        tf.keras.layers.Rescaling(1./255, input_shape=(300,300,3)),
        tf.keras.layers.RandomRotation(0.11,fill_mode='nearest'),
        tf.keras.layers.RandomTranslation(0.2,0.2,fill_mode ='nearest'),
        tf.keras.layers.RandomZoom(0.2,fill_mode='nearest'),
        tf.keras.layers.RandomFlip('horizontal'),
        tf.keras.layers.Conv2D(conv_n*1, conv_s, activation = h_act),
        tf.keras.layers.MaxPooling2D(mp_s),
        tf.keras.layers.Conv2D(conv_n*2, conv_s, activation = h_act),
        tf.keras.layers.MaxPooling2D(mp_s),
        tf.keras.layers.Conv2D(conv_n*4, conv_s, activation = h_act),
        tf.keras.layers.MaxPooling2D(mp_s),
        tf.keras.layers.Conv2D(conv_n*4, conv_s, activation = h_act),
        tf.keras.layers.MaxPooling2D(mp_s),
        tf.keras.layers.Conv2D(conv_n*4, conv_s, activation = h_act),
        tf.keras.layers.MaxPooling2D(mp_s),
        
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(512, activation=h_act),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer ='Adam', loss ='binary_crossentropy', metrics =['accuracy'])
    
    ## 추출 단계 (Extraction)
    data =tfds.load('horses_or_humans',split='train',as_supervised =True)
    val_data  =tfds.load('horses_or_humans', split= 'train', as_supervised=True)
    
    ## 변환단계 (Transform)
    # -> model layer 에서 이루어짐으로 함수 작성필요없음 addon service 가끝나서 이렇게 해야함. 
    train_batches = train.shuffle(100).batch(32)
    validataion_batches = val_data.batch(32)
    
    ## 적재 단계(load )
    history = model.fit(train_batches,epochs= 3, validation_data = validataion_batches)
    return model 
model =ETL_process()
```

## 4.6.2  ETL 병렬화

```python
train_data = tfds.load('cats_vs_dogs',split='train[:10%]',with_info=True)

def file_extraction_from_tfds(data_name ='cats_vs_dogs'):
    file_pattern = os.path.join(
        os.path.expanduser("~")+
        f'/tensorflow_datasets/{data_name}'
    )
    def listdir_show(file_pattern):
        for order,i in enumerate(os.listdir(file_pattern)):
            # print(order,i)
            if order<2:
                version = i
                
        return version
    version = listdir_show(file_pattern)
    file_pattern =os.path.join(file_pattern,version)
    # listdir_show(file_pattern)
    from glob import glob
    file_pattern = glob(os.path.join(file_pattern, f'{data_name}-train.tfrecord*'))[0]  
    file =tf.data.Dataset.list_files(file_pattern)
    print(file_extraction_from_tfds)
    return file
files =file_extraction_from_tfds()
train_dataset = files.interleave(
    tf.data.TFRecordDataset,
    cycle_length= 4, # 동시에 전처리할 입력 원소 개수 지정
    num_parallel_calls=tf.data.experimental.AUTOTUNE ## 병렬 실행 횟수 지정. cpu 개수에 따라 동적으로 설정
    
)
print(train_dataset)

```

```python
## 레코드를 로드할때 디코딩 하는 매핑 함수 
def read_tfrecord(serialized_example):
    feature_description={
        'image': tf.io.FixedLenFeature((), tf.string,""),
        'label': tf.io.FixedLenFeature((), tf.int64,-1)
    }
    example = tf.io.parse_single_example(
        serialized_example, feature_description
    )
    image = tf.io.decode_jpeg(example['image'], channels= 3)
    image = tf.cast(image, tf.float32)
    image = image/255
    image = tf.image.resize(image, (300,300))
    return image, example['label']
```

## 병렬화 -> ㅡmultiprocessing cpu 8개 

```python
import multiprocessing
cores = multiprocessing.cpu_count()
print(cores)
train_dataset = train_dataset.map(read_tfrecord, num_parallel_calls=cores)
train_dataset = train_dataset.cache() # 메모리에 데이터셋을 캐싱 
```
