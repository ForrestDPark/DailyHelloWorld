> ### fashion mnist data 러닝

```python
(training_images, training_labels),(test_images, test_labels)=\
    tfds.as_numpy(tfds.load('fashion_mnist',
                            split  =['train','test'], ## split 을 하지않으면 dic 반환,
                            ## split 변수를 지정하면 분할 세트의 리스트가 반환됨. 
                            batch_size = -1, ## tf.Tensor 타입의 전체 데이터 반화
                            as_supervised =True 
                            ))


def model_training_for_tfds(training_images, training_labels,test_images, test_labels):
    ## 2024.10.17
    ## fashion mnist 인 경우 tf.unit8 타입의 텐서가 반환되기떄문에  cast 를 통해 실수로 변환후 255로 나누어야함. 
    training_images = tf.cast(training_images,tf.float32)/255.0
    test_images = tf.cast(test_images,tf.float32)/255.0
    
    model = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(input_shape =(28,28,1)),
        tf.keras.layers.Dense(128, activation=tf.nn.relu),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(10,activation=tf.nn.softmax)
        
    ])
    model.compile(optimizer = 'adam',
                  loss = 'sparse_categorical_crossentropy',
                  metrics = ['accuracy'])
    model.fit(training_images,training_labels, epochs=5)
model_training_for_tfds(training_images, training_labels,test_images, test_labels)
```
