```python
## @ 2024/05.22 
#오븐시계 a
current_hour, current_min  = map(int, input().split())
cook_time = int(input())

# 오븐구이가 끝나는 시간? 계산 
# 14 :  30 + 80 min

#1. cook time 정제 
cook_hour, cook_min  =  divmod(cook_time,60)  

finish_hour = (current_hour + cook_hour) % 24
finish_min =  current_min + cook_min
## min  이 60 분을  넘었을 때 
if finish_min >= 60:
    finish_hour = (finish_hour + divmod(finish_min,60)[0]) % 24
    finish_min = divmod(finish_min,60)[1]

print(finish_hour, finish_min)
    
```

```python
## 세개의 주사의 상금 계산 프로그램

a,b,c = map(int, input().split())
if a==b==c:
    reward =10000 + a*1000
elif a==b or b==c :
    reward = 1000+b*100
elif a==c :
    reward = 1000+a*100
else :
    reward = max([a,b,c])*100
    
print(reward)



```
