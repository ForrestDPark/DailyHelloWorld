```python


def seleniumSettings():
    from selenium import webdriver
    from selenium.webdriver.chrome.service import Service
    from webdriver_manager.chrome import ChromeDriverManager
    from selenium.webdriver.common.by import By
    
    ## Chrome version & driver version check
    chrome_options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_options)
    driver.get("https://www.jobkorea.co.kr/")
    # 현재화면 Capture 하기
    driver.save_screenshot("../Data/001.png")
    
seleniumSettings()
```

```python
def detail_page_crawling(di_url):
        from bs4 import BeautifulSoup
        import urllib.request as req
        from urllib import parse
        import time 
        def printContent(title ,tblist):
                # print(tblist[0])
                if tblist :
                        for i, c in enumerate(tblist): ## 경력,학력, 고용형태 
                                if i<1:
                                        print(f"{title}({i}) :",(c.get_text().replace(" ","").replace("\r\n"," ").replace("\n", " ").strip()),sep="")
                                elif i>3:
                                        print("...")
                                        break
                                        
                                else: print(f"   |-+ etc :",(c.get_text().replace(" ","").replace("\r\n"," ").replace("\n", " ").strip()),sep="")
                        
        detail_url = parse.quote(di_url, safe=":/?=&")
        print(" |- 채용링크:",detail_url)
        
        res = req.urlopen(detail_url)
        soup = BeautifulSoup(res, "html.parser")
        contents = soup.select("#container > section dl ")
        skil =soup.select("#container > section  div:nth-child(1) > dl > dd:nth-child(6)")
        carear =soup.find_all(class_="col_1")
        tblist =soup.find_all(class_="tbList")
        pref = soup.find_all(class_="pref")
        pay = soup.select("#container > section div:nth-child(2) > dl > dd:nth-child(4)")
        loc = soup.select("#container > section  div:nth-child(2) > dl > dd:nth-child(6) > a")
        corpInfoLink = "https://www.jobkorea.co.kr"+soup.find("div", class_="coBtn").find("a").get("href") if soup.find("div", class_="coBtn") else ""
      
        workTime = soup.select("#container div:nth-child(2) > dl > dd:nth-child(8)")
        # print(tblist)
        print(" |- 기업정보링크:",corpInfoLink)
        printContent(" |- 급여",pay)
        printContent(" |- 근무지",loc)
        if skil :
                print(" |- 스킬:",skil[0].get_text().replace("\r\n"," ").replace("\n"," ").strip())
        if carear:
                for i, c in enumerate(carear): ## 경력,학력, 고용형태 
                        if i < 3:
                                print(f' |- {["경력","학력", "고용형태" ][i]} : {c.get_text()}')
                        else:
                                print(f' |- 기타 : {c.get_text()}')
        printContent(" |- 시간",workTime)
        printContent(" |- 우대조건",pref)
        
                
detail_page_crawling("https://www.jobkorea.co.kr/Recruit/GI_Read/44665161?Oem_Code=C1&logpath=1&stext=%EA%B0%9C%EB%B0%9C&listno=1")
```
