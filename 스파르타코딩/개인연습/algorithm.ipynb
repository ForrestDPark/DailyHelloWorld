## 문제 005 나머지 합 구하기 
N개의 수 A 1 , A 2 , ..., A N 이 주어졌을 때 연속된 부분의 합이 M으로 나누어떨어지는 구간의 개수를 구하는 프로그램을 작성하시오. 즉, Ai + ... + Aj (i ≤ j)의 합이 M으로 나누어떨어지는 (i, j) 쌍의 개수를 구하 시오.


ㅁㄴ

```python
import numpy as np

sample=[1,2,3,4,5,56,34,5]
#(i,j) ->

def sum_ofRange(i,j,sample,M):
    count =0
    for i in range(len(sample)):
        for j in range(len(sample)):
            if i<j and np.sum(sample[i:j])%M==0:
                #print(f"sum({i}:{j}) = {np.sum(sample[i:j])}")
                count+=1
                #print(f"mode(sum({i}:{j}),{M}) = {np.sum(sample[i:j])%M}")
    print("개수 :",count)

sum_ofRange(1,2,sample,3)
```
