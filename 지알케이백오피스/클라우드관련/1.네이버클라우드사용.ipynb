### 2. 네이버 클라우드 장점 
- 3개월 무료 사용, 10만원 크래딧,  공공 기관 서비스ㅅ 네이버 클라우드 필수 
- vpc 서비스 제공(보안강화)
- 금융, 공공, 일반 서비스로 분류하여 사용 
- aws 처럼 여러 서비스 지원(왠만한 것 다있음)
- 라이브 스트리밍, 쿠버네티스 등 지원 

### 3. NCP - 개인 계정으로 가입하기 
- 전화번호는 1개만 등록 가능 
- 결재 카드 등록 무료 크레딧 10만원 (3개월 이내 사용 필요)
- 콘솔 클릭 -> 좌 사이드바 에서 VPC(virtual private cloud) 인지 확인 -> service -> vpc 클릭->

### 4. 네트워크 세팅 서브넷 생성 및 설정 
#### 4.1 VPC 생성
- vpc 를 생성한다. service 명  back-office-test , 10.0.0.0/22 = C class 9 개의 네트워크 사용 
- subnet 이름 vpc 그룹, ip 주소 범위 (⭐️ 10.0.1.0/24), 
- 가용존( KR-2 선택, kr 2 가 나중에 나온 것)
- 서버  생성 콘솔 선택 :
    기존 콘솔 화면, 신규 콘솔 화면 둘중 신규콘솔화면 선택(더많은 기능 사용 가능)
    1. 빠르고 쉬운 서버ㅣ 이미지 선택,하이퍼 바이저 별 서버 이미지 제공, 서버 스펙코드로 서버사양 쉽게 설정, 추가 스토리지 생성 기능 제공 ( kvm 기본 스토리지 10G ~ 2TB 까지 자유롭게 생성 , 추가 스토리지 20개 가능. XEN 기존 제공 스토리지 스펙과 동일)

### 6. 서버 세팅
서버 이미지 ubuntu-22.04 체크 -> 다음
- vpc, subnet, server 스펙 -> micro 서버 타입 선택
    : 신규강비후 최초 결제 등록 한 월 부터 1년동안 무료 제공, 무료기간 지난후 요금 자동 부과 2025.5 월말 까ㅣ 죵금 발생 없음. 
- micro, mi1-g3(vCPU 1EA, memory 1GB ) 체크 
- network interface +추가 버튼 클릭, (IP 자동할당)
- 공인 ip 설정 ( 월 이용료 4032 원) 사용하지 않을때는 반납 필. 
- 스토리지 설정 ( 10GB 이하는 무료 임 )
- 인증키 설정 -> 다운 -> 관리 폴더에 보관 
- 네트워크 접근설정 (
     보유 ACG (Access control group: 방화벽 구축 없이 서버 크룹에 대한 접근제어 관리 상품) 선택 
- server 생성 버튼 클릭 -> 수분 수십분 소요
