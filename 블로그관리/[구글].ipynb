# [글쓰기] 나만의 글쓰기 순서 

글쓰는 순서를 그냥 외부의 정보를 마구 짜집기로 수집하면서 메모형식으로 틀에 모아둘때는 마크다운이 오히려 편하다. 
그리고 수집한 정보에서 바로 글을 만드는 것이아니라 논리도해를 만든다. 이 작업은 프레젠테이션에서 모형과 선만으로 충분하다. 
도해를 다 짜고 나면 전체 글의 맥락이 잡히고 정리된다. 이 그림을 하나씩 마크다운에 입혀서 글의 완성도를 높이는 작업을한다. 

파워포인트에는 레이아웃을 지정해놓고 내가 원하는 레이아웃을 바로 적용해서 프레젠테이션할 수있는 점이 편리하다.
