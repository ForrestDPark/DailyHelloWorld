# **<span style="color:#21409a"> 🤔 Stable Diffusion이란?</span>**

> ### ✅ <span style="color: #21409a">** Stable Diffusion**</span> 는 <span style="background-color: #fff5b1"><span style="color: #008080">**무작위 노이즈 이미지 로부터 역방향 확산을 통해 원하는 이미지를 생성하는 방식**</span></span> 이다.

- Stability AI의 공식 파트너사 AICU Inc. 에서 제공
- **Stable Diffusion XL(SDXL)** 모델을 사용함. 
- AUTOMATIC1111/Stable Diffusion WebUI(v1.8.0)
- 디지털 일러스트레이션 제작할 때 쓰임
- Google colab 활용 가능함. <br>
- **SDXL**에는 OpenAI 가 개발한 <u>**CLIP-ViT/L**</u> 과 오픈소스인 OpenCLIP-ViT/G , 테스트 데이터인 **LAION** 이 적용됨
- **잠재확산 모델(latent diffusion model)**: 잠재공간(latent spcae)으로 압축 하여 모델 학습, 노이즈 생성함. 
- Stable Diffusion 은  <u>**CLIP-ViT/L**</u>의 **텍스트 인코더 부분**을 활용해 프롬프트를 해석함. <br>

# **<span style="color:#21409a"> 🤔 U-Net 이란?</span>**

# 디퓨전 모델이란?
- 생성형 모델, 
- 학습받은것과 비슷한 새로운 데이터를 생성하도록 설계된 모델 
- 순방향 확산(Forward Diffusion)
- 학습용 이미지에 점차적으로 잡음 추가, 점점 아무런 특징이 없는 무작위잡음(random noise) 이미지로 바꿈. 
- 물컵에 잉크 한방울 떨어뜨리는 것처럼 처음에 이크가 어디에 떨어졌는지 전혀 알수없게 됨. 
  
- 역방향 확산 (Reverse diffusion)
- 의미 없는 이미지에서 출발해 의미 잇는 이미지로 복구됨. 잉크물에서 잉크가 떨어진 위치를 찾아냄.(drift)
  
1. 이미지에 얼마나 많은 잡음이 들어있는지 알아야함. 
2. NN 모델에 추가된 잡음 예측 noise predictor(U-Net)사용 

[학습]
1. 학습용 이미지 선택 
2. 무작위 잡음 이미지 생성
3. 학습용 이미지에 정해진 단계만큼 무작위 잡음 이미지를 추가하여 손상 
4. 잡음 예측기에 잡음 추가량을 학습시킴. 가중치조정 후 정답을 보여주는 방식 작업 수행

[예측] 역방향 확산
1. 완전한 잡음 이미지 생성 
2. 잡음 예측기에게 이미지에 잡음이 얼마나 포함되었냐는 질문
3. 잡음 예측기가 반환한 예측 잡음을 원래의 잡음 이미지에서 제(substract)
4. 반복 (조건 부여없이 이미지가 생성됨)

이미지공간의 크기 512*512 786462 차원.
너무 큼. DALL-E 는 이런 이미지 공간에서 작동. 한계가 있음. 



잠재확산모델(latent diffusion model)
- 먼저 잠재공간으로 이미지를 압축한뒤 연산 실행. 
- 잠재 공간은 이미지 공간에 비해 48배작음. 


가변 자동인코더(Variatinal AutoEncoder,VAE)
- 이미지를 잠재공간(낮은차원)으로 압축하는데 사용함. 
- 디코더가 잠재공간으로부터 이미지를 복원함. 
- 잠재공간 크기 4x64x64(16,384) 
- 잠재공간상의 무작위텐서(random tensor, latent noise) 를 생성. 
- 이미지 손상시킬때 공간상의 이미지 표현을 잠재잡음을 사용해 손상시켜 연산을 훨씬 빠르게함.
