# 읽기 쉬운 글을 쓰는 방법은 무엇일까? CS-GPS 란?
<!--💡이글을 쓰게 된 이유는 무엇인가요?-->
💡 읽기 쉬운 글에 대한 수요는 많고 이에 대해 알려주는 잡다한 기술들이 많다. 하지만 본질적인 방법은 글의 구조와 체계를 잡는 훈련하는 것이다. 범주, 요점, 그룹, 병렬, 순서 를 통해 쉬운글의 체계를 알아보자. 
***
전한다 : 특정 주제에 관해 정보를 정리하고 표현하여 수신자에게 전달하는것. 
주제 :  다양한 구성요소가 일정한 구조로 연결된 현실 세계의 무언가. 


요약은 중요한 점을 짧게 표현한 것이다. 

요약은 범주와 요점을 생각한다. 
범주는 항목에 해당하는 정보이고 요점은 항목을 정리한 내용. 


분류 병렬 순서의 구조 찾기
- 구조화 : 분류, 병렬, 순서 
분류(G) : 요소에서 공통점을 찾아 그룹으로 정리.공통점을 나타내는 이름 짓기.
병렬(P) : 필연적으로 표형식의 구조를 갖게 됨.(항목에 대한 대응 구조로 비교가능.)
순서(S) : 여러 대상의 순서를 정한 구조. 

범주, 요점 (Category & Summary,CS)

보고서는 CS 와 GPS 를 정리한다. 

CS-GPS

1. 먼저, 그룹이 될 부분을 찾는다. 
2. 그룹이 여러개라면 이를 조합하여 표 형식으로 만들 수는 없는지 생각한다.
3. 그룹에 속한 요소에 순서를 매길 기준은 없는지 찾는다. 


범주와 요점

A : 1
B : 2
C : 3 

범주 : 요점 

패턴 : 몇번씩 등장하는 같은 형태.  C또는 G 에서 패턴인식이 가능.

# **<span style="color:#21409a"> 🚶‍♂️정보정리의 순서</span>**
## &emsp;(1)**관련 정보를 모두 수집**하고 먼저 **정리되지 않은 글을 작성**한다.<br>
## &emsp;(2)정보속에서 **반복적으로 등장하는 요소**에 주목하여 **그룹화**한다. 
## &emsp;(3)논리도해를 통해 **그룹을 묶고 명칭과 행동(동사)를 표현**한다.<br>
### &emsp; &emsp;**✅ 손으로 그리거나 파워포인트로 그린다.**
## &emsp;(4)정보의 **패턴을 파악하여 체계화**(매뉴얼화) 한다.<br>
## &emsp;(5)긴글을 **주제별로 분절**하고 **순서가 3개를 넘지않게 분류**한다<br>
- 분절예시 -> [사건개요]...[관계자]...[올바른 절차]...
- 분류예시 -> 필요한장비 (1),(2),(3)...설치 순서 (1),(2),(3)...
