   (2) DNS 설정 클릭 > CNAME HTTPS 인증서 정보 입력 
   (3) 타입에 CNaME 선택 > http 인정서 구매후받은 Host 값 입력 > Value 값을 복사해 값/위치 에 입력 후 마자미가에 "." 을 붙여 줌. ex) ValueCode.codecode.comomoo.com.
   (4) 확인 > 저장 > https 인증 설정 완료

**`인증기관 선택시 주의사항 `**
1. 신뢰성 및 브라우저 호환성 
2. 인증서 갱신 및 자동화 
3. 서버 보안 (패치 업데이트 적용)


**`SSL/TLS 종류`**
|종류|설명|
|--|--|
|DV(Domain Validation)|도메인 소유권 만 확인하는 가장 기본적 검증|
|OVl(Organization Validation)|기업,조직에 대한 정보 확인 , 발급기관이 조직신뢰성 검증함|
|EV(Extended Validation)|가장 높은 수준의 검증, 조직 신뢰성, 법적존재여부 등 추가 검증|

SSL 인증서 적용 절차 
(1) 인증서 신청 > (2) 인증 > (3) 인증서 발급 > (4) 서버 인증서 설치 단계


**`DV(Domain Validated) `**
- 실제 해동 도메인의 관리자가 맞는지 확인을 거치는 인증방법 
- 메일, dns, 웹 인증 이있음.
1) 메일 인증 : 수신 할 수 있는 메일 주소가 정해져 있음 도메인 관리자 정보에 등록된 메일 혹은 admin@신청도메인 처럼 발급사에서 공식으로 인정하는 메일 주소만 가능 
2) DNS 인증 : 도메인 네임서버 zone 파일에 발급사가 제공하는 인증 레코드 값을 입력하는 방법. 
    -zone 파일 : 자체적으로 네임 서버를 구축한게 아닌경우 네임서버 이용업체(가비아등)에 레코드 값 입력방법 문의 
3) 웹 인증 : 발급사 제공 txt 파일을 도메인에 연결된 서버에 업로드하는 방법
